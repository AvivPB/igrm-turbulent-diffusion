```python
for prop, log_prop, xrange in zip(props_to_plot, log_props, ranges):
    print(prop)
    lw = len(caesar_dict.keys())*0.5
    for sim, info in caesar_dict.items():
        # if 'snap' not in info:
        #     continue
        values = info['values'][prop]
        values_to_plot = values
        print(len(values_to_plot))
        # print(values_to_plot)
        if log_prop:
            nzeros = np.count_nonzero(values_to_plot==0)
            print(f'nzeros={nzeros}')
            values_to_plot = np.log10(values_to_plot)
        # print(values_to_plot)
        values_to_plot = values_to_plot[np.isfinite(values_to_plot)]
        print(len(values_to_plot))
        if log_prop:
            try:
                zero_vals_to_plot = [np.min(values_to_plot)-ndex_below_min]*nzeros
                # print(len(zero_vals_to_plot))
                values_to_plot = np.append(values_to_plot, zero_vals_to_plot)
            except Exception as e:
                print(e)
                values_to_plot = np.append(values_to_plot, zero_vals_to_plot)
        print(len(values_to_plot))
        # print(values_to_plot)
        print()
        plt.hist(values_to_plot, bins=40, range=xrange, log=True,
                 histtype='step', ls='--', linewidth=lw,
                 label=f'{sim}: z={info['z'][0]:.2f}')
        lw -= 1*0.5
        # print(lw)
    xlabel = rf'{prop} [{values.units}]'
    if log_prop:
        xlabel = f'log({xlabel})'
    plt.xlabel(xlabel)
    plt.ylabel('Counts')
    plt.legend(fontsize=6, loc=(0,1.02))
    plt.grid()
    fig = plt.gcf()
    fig.set_dpi(200)
    plt.show()
```
