# Plot properties over time

$N_\mathrm{BH}-a$

```python
color_list = ['C0', 'C1', 'C4', 'C6']
size_list = [1e-2, 1e-2, 1e-2, 1e-2]
alpha_list = [0.7, 0.7, 0.7, 0.7]

handles = []
for i, (sim, sim_info) in enumerate(sim_dict.items()):
    plt.scatter(sim_info['info']['sim_time'], sim_info['info']['nbh'],
                color=color_list[i], s=size_list[i], alpha=alpha_list[i],
                label=sim)
    handle= Line2D([], [], color=color_list[i], ls='',
                   marker='.', ms=3, alpha=1, mfc=color_list[i], label=sim)
    handles.append(handle)
    
# plt.scatter(sim_time, nbh, s=1e-3)

plt.legend(handles=handles, loc=(0, 1.02), fontsize=8)

plt.xscale('log')
plt.yscale('log')

plt.xlabel('a')
plt.ylabel(r'$N_{\mathrm{BH}}$')

fig = plt.gcf()
fig.set_dpi(200)
```
